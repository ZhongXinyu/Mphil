```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set the true gradient to be 0.5

m = 0.5

# add noise to the gradient

m1 = m + np.random.normal(0, 0.1)

m2 = m + np.random.normal(0, 0.1)

# generate random intercepts

c1 = np.random.normal(0, 1)

c2 = np.random.normal(0, 1)

```

```python
# generate the size of the dataset

n1 = 100

n2 = 200
```

```python
# generate 100 random x values

x1 = np.random.uniform(0, 10, n1)

x2 = np.random.uniform(0, 10, n2)

# generate 100 random y values with the true gradient and random intercepts

y1 = m * x1 + c1

y2 = m * x2 + c2

# add noise to the y values

y1 = y1 + np.random.normal(0, 0.1, n1)

y2 = y2 + np.random.normal(0, 0.1, n2)
```

```python
# plot the data

plt.scatter(x1, y1, color='red')

plt.scatter(x2, y2, color='blue')

plt.show()
```

```python
# calculate gradient of first dataset
```

```python
# calculate the estimated gradient

m_est = (n1 * np.sum(x1 * y1) + n2 * np.sum(x2 * y2) - (np.sum(x1) * np.sum(y1)) - (np.sum(x2)* np.sum(y2))) / (n1 * np.sum(x1 * x1) - (np.sum(x1) **2) + n2 * np.sum(x2 * x2)- (np.sum(x2) **2))
print ("Estimated gradient is: ", m_est)
print ("True gradient is: ", m)
```

```python
# calculate the estimated intercept

c1_est = (np.sum(y1) - m_est * np.sum(x1)) / n1

c2_est = (np.sum(y2) - m_est * np.sum(x2)) / n2

print ("Estimated intercepts are: ", c1_est, c2_est)
print ("True intercepts are: ", c1, c2)
```

```python
# plot the data and the estimated line

plt.scatter(x1, y1, color='red')

plt.scatter(x2, y2, color='blue')

plt.plot([0, 10], [c1_est, m_est * 10 + c1_est], color='black')

plt.plot([0, 10], [c2_est, m_est * 10 + c2_est], color='black')

plt.show()
```
