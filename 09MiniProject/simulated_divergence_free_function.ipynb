```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
def fn(x1,x2,a):
    return np.array(
        [
            np.exp(-a*x1*x2)*(a*x1*np.sin(x1*x2)-x1*np.cos(x1*x2)),
            np.exp(-a*x1*x2)*(x2*np.cos(x1*x2)-a*x2*np.sin(x1*x2))
        ]
        )
```

```python
# generate a grid of points
x1 = np.linspace(0, 4, 50)
x2 = np.linspace(0, 4, 50)
x = np.meshgrid(x1,x2)

# evaluate the function at each point
a = - 0.01
y = fn(x[0],x[1],a)

# Plot
plt.quiver(x1, x2, y[0], y[1], angles='xy', scale_units='xy', scale=20, color='r')

# Add labels and title
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Vector Field Plot')

# Set the limits of the plot
plt.xlim(0, 4)
plt.ylim(0, 4)

# Show plot
# plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
# save the data as in a json file
# 2500 data points with 2 features and 2 labels
data = {
    "x1":x[0].flatten().tolist(),
    "x2":x[1].flatten().tolist(),
    "y1":y[0].flatten().tolist(),
    "y2":y[1].flatten().tolist()
}
with open("testdata/vector_field.json", "w") as outfile:
    json.dump(data, outfile)
```
