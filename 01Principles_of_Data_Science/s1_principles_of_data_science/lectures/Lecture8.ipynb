## Convergence in distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
plt.style.use('../mphil.mplstyle')

fig, ax = plt.subplots(1, 3, figsize=(19.2,4.8))

x = np.linspace(-3,3,200)

v = np.random.normal(size=100)
ax[0].hist(v, bins=50, range=(-3,3), density=True)
ax[0].plot(x, norm.pdf(x) ,lw=3)
ax[0].text(0.02,0.98,f'N=100', transform=ax[0].transAxes, va='top')

v = np.random.normal(size=1000)
ax[1].hist(v, bins=50, range=(-3,3), density=True)
ax[1].plot(x, norm.pdf(x) ,lw=3)
ax[1].text(0.02,0.98,f'N=1K', transform=ax[1].transAxes, va='top')


v = np.random.normal(size=1000000)
ax[2].hist(v, bins=50, range=(-3,3), density=True)
ax[2].plot(x, norm.pdf(x) ,lw=3)
ax[2].text(0.02,0.98,f'N=1M', transform=ax[2].transAxes, va='top')

```

## The central limit theorem

- For X_i generated from <font color="red">*any*</font> distribution with means, $\mu_i$, and variances, $\sigma_i^2$ then the sum

$$ S = \sum_i X_i $$

$\quad \;$ has mean and variance 

$$ \mu = \sum_i \mu_i \quad \text{ and } \quad \sigma^2 = \sum_i \sigma_i^2 $$

- The CLT states that the distribution of $S$ will tend to a normal distribution, $G(\mu, \sigma)$ as $N\to\infty$

- Any distribution with defined mean and variance and large $N$ tends to the normal distribution
